```python
def round_nearest(x, a):
    return round(x / a) * a

df["Tip_percentage_1"] = (df["Tip_amount"]/df["Fare_amount"]).fillna(0) 
df["Tip_percentage_2"] = (df["Tip_amount"]/round_nearest(df["Total_amount"]-df["Tip_amount"],0.05)).fillna(0) 
df["Tip_percentage_1"] = (1.25*df["Total_amount"]-df["Tip_amount"]).fillna(0) 

#df["Tip_percentage_2"] = (df["Tip_amount"]/(df["Total_amount"]-df["Tip_amount"]-df[""])).fillna(0) 

plt.figure(figsize=(12,8))
v1 = df.loc[df["vendorid"]==1]
v1 = v1.loc[(v1["Payment_type"]==1)] 
v2 = df.loc[df["vendorid"]==2]
v2 = v2.loc[(v2["Payment_type"]==1)]

# Did the rounding system change at some time during the year? Check this.
# v1 = v1.loc[v1["pickup_datetime"].dt.month > 11]
# v2 = v2.loc[v2["pickup_datetime"].dt.month > 11]

v1["Tip_percentage_1"].loc[v1["Tip_percentage_1"]<1].hist(bins=np.arange(0.19,0.31,0.0005))
plt.show()
plt.figure(figsize=(12,8))
v2["Tip_percentage_1"].loc[v2["Tip_percentage_1"]<1].hist(bins=np.arange(0.19,0.31,0.0005))
plt.show()
```

# Time of day trends

```python
df["Day"] = pd.to_datetime(df["pickup_datetime"]).dt.dayofweek

df["dropoff_datetime"] = pd.to_datetime(df["dropoff_datetime"])
df["pickup_datetime"] =  pd.to_datetime(df["pickup_datetime"])     
df["Trip_time"] = df["dropoff_datetime"] - df["pickup_datetime"]
df["Trip_time"] = df["Trip_time"] / pd.Timedelta('1 minute')


df["Time"] = pd.to_datetime(df["pickup_datetime"]).dt.time
```

```python
temp = df.loc[df["Payment_type"]==1]
temp = temp.loc[df["Tip_percentage_2"]<1]
temp = temp.loc[df["Tip_percentage_2"]>0]
weekdays=temp.loc[df["Day"]<5]
weekdays.groupby(weekdays["pickup_datetime"].dt.hour).agg("mean")["Tip_percentage_2"].plot(kind="bar")
plt.ylim([0.19,0.21])
```

```python
weekends=df.loc[df["Day"]==5]
weekends = weekends.loc[weekends["Tip_percentage_2"]<1]
weekends = weekends.loc[weekends["Tip_percentage_2"]>0]
weekends.groupby(weekends["pickup_datetime"].dt.hour).agg("mean")["Tip_percentage_2"].plot(kind="bar")
plt.ylim([0.19,0.205])
```

```python
weekends=df.loc[df["Day"]==6]
weekends = weekends.loc[weekends["Tip_percentage_2"]<1]
weekends = weekends.loc[weekends["Tip_percentage_2"]>0]
weekends.groupby(weekends["pickup_datetime"].dt.hour).agg("mean")["Tip_percentage_2"].plot(kind="bar")
plt.ylim([0.19,0.205])
```
