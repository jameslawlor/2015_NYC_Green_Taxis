# Geohash

It is too computationally intensive to calculate which neighbourhood each unique ~20 million lat/lons of the pickup locations in the taxi dataset correspond to. To circumvent this we will "geohash" or encode each set of coordinates to a rougher precision, and then decode these back to lat/lon values. This results in around ~6000 unique points which is far easier to handle

```python
# Geohash magic
import geohash

df["coords"] = list(zip(df["Pickup_latitude"],df["Pickup_longitude"]))
df["coords"] = df["coords"].apply(lambda x : geohash.encode(x[0],x[1],6))
df["coords"] = df["coords"].apply(geohash.decode)
```

```python
unique_positions = df["coords"].unique() # List of the decoded ~6000 unique positions
```

```python
from shapely.geometry import Point, shape

def neighbourhood_finder(coords):
    """Returns the NYC neighourhood from an input set of coordinates"""
    lat = coords[1]
    lon = coords[0]
    point = Point(lat,lon) 
    for feature in nyc['features']: # Iterate through all shapes in the NYC shape file
        polygon = shape(feature['geometry'])   
        if polygon.contains(point):
            return feature["properties"]["neighborhood"]        

# Create a dictionary of lat/lon : neighbourhood mappings
d2 = {}
for pos in unique_positions:
    d2[pos] = neighbourhood_finder(pos)
```

```python
df["neighbourhood"] = df["coords"].apply(lambda x : d2[x]) # apply lat/lon->neighbourhood mapping from above
```

```python
# Count number of pickups in each neighbourhood
count_dict = dict(df.groupby("neighbourhood").agg("count")["coords"])
```

# Mapping
We will use the above aggregation to produce a choropleth map of pickups by neighbourhood using the Python library Bokeh

## Wrangle the neighbourhood data into a bokeh-friendly format

```python
good_dic = {}
for hood in list(d.keys()):
    try:
        good_dic[hood] = {}
        good_dic[hood]["count"] = count_dict[hood]
        good_dic[hood]["xs"] = d[hood]["xs"]
        good_dic[hood]["ys"] = d[hood]["ys"]
    except:
        pass
# for item in good_dic:
#     print(len(good_dic[item]))
      
good_dic = {k:v for k,v in good_dic.items() if len(good_dic[k])>1}
# print(len(good_dic))
```

```python
from bokeh.io import show
from bokeh.models import (
    ColumnDataSource,
    HoverTool,
    LogColorMapper
)
from bokeh.plotting import figure
from bokeh.models import HoverTool
from bokeh.plotting import figure, show, output_file, ColumnDataSource
from bokeh.io import output_notebook

output_notebook()

# Prepare data for choropleth
hood_xs = [v["xs"] for k,v in good_dic.items()] # x-values
hood_ys = [v["ys"] for k,v in good_dic.items()] # y-values
hood_counts = [v["count"] for k,v in good_dic.items()] # Counts by neighbourhood
hood_names = [k for k,v in good_dic.items()] # Names of neighbourhoods

# Produce a nice colour scheme here
import seaborn as sns
colors = sns.color_palette("Reds",n_colors=50).as_hex()

# List of each neighbourhood's colour based on it's count. Use a natural log transform for aesthetics
hood_colors = [colors[int((len(colors)-1)*np.log1p(count)/max(np.log1p(hood_counts)))] for count in hood_counts]

#Prepare for bokeh
source = ColumnDataSource(data=dict(
    x=hood_xs,
    y=hood_ys,
    color=hood_colors,
#    color="Blue",
    name=hood_names,
    count=hood_counts,
))
```

```python
# Interactive tools
TOOLS="pan,wheel_zoom,box_zoom,reset,hover,save"

# Initialise figure
p = figure(title="2015 NYC Green Taxi Pickup by Neighbourhood", tools=TOOLS,width=1200,height=900)

# Initiatlise Neighbourhood shapes 
p.patches('x', 'y', source=source,
          fill_color='color', fill_alpha=0.7,
          line_color="white", line_width=0.5,legend="test")

# Create hovertool + interactivity
hover = p.select_one(HoverTool)
hover.point_policy = "follow_mouse"
hover.tooltips = [
    ("Name", "@name"),
    ("Counts)", "@count"),
    ("(Long, Lat)", "($x, $y)"),
]

output_file("2015_green_taxis.html", title="2015 NYC Green Taxi Pickups by Neighbourhood")

show(p)
```
